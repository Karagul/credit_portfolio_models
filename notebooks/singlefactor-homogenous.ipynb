```python
default_probability = 0.05
correlation = 0.2 
loss_given_default = 1.0
x_list = np.arange(0.0, 1.00, 0.01)
loss_f_values = []
loss_d_values = []
```

```python
for i in x_list:
    loss_f_values.append((i, portfolio.loss_funtion(i, default_probability, correlation, loss_given_default))) 
    loss_d_values.append((i, portfolio.loss_density(i, default_probability, correlation, loss_given_default)))
data = [{'name': 'loss function', 'data':loss_f_values},
        {'name': 'loss density' , 'data':loss_d_values},]
```

```python
import highcharts_lines as chart
reload(chart)
chart.initialize_highcharts()
chart.create_from_data_series(data, title = 'ASRF approximation', subtitle = "with different loss percentage",
                              ylabel = ' value ', xlabel = ' loss ', xtype = 'string')
```
